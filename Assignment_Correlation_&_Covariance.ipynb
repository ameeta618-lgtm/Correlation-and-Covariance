Assignment : Correlation & Covariance

Q1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

    ->Correlation measures both direction and strength of relationship and is scaled between -1 and +1.
    Covariance depends on units while correlation is unit-free.

Q2. What does a positive, negative, and zero covariance indicate about the relationship between two variables?

    ->Positive covariance: variables move in same direction.
    Negative covariance: variables move in opposite direction.
    Zero covariance: no linear relationship.

Q3. Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?

    ->Limitations of covariance:
    - Value depends on scale of measurement.
    -Hard to interpret magnitude.
    Correlation is preferred because it standardizes the value between -1 and +1.

Q4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation coefficient. When would you prefer to use Spearman’s correlation?

    ->Pearson correlation measures linear relationship between continuous variables.
    Spearman correlation measures rank-based (monotonic) relationship.
    Spearman is preferred when data is ordinal or not normally distributed.

Q5. If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you infer causation from this value? Why or why not?

    ->If correlation = 0.85, it indicates strong positive relationship.
    However, correlation does not imply causation because other variables may influence the
    relationship.
